```python
fuel= pd.get_dummies(auto['Fuel Type'], drop_first = False)
fuel
```

```python
auto = auto.drop(['N','D'],axis = 1)
auto
```

```python
auto
```

```python
from sklearn.model_selection import train_test_split
# We specify this so that the train and test data set always have the same rows, respectively
df_train, df_test = train_test_split(auto, train_size = 0.85, test_size = 0.15, random_state = 1)
```

```python
y_train=df_train['CO2 Emissions(g/km)']
x_train=df_train.drop(['CO2 Emissions(g/km)'],axis=1)
```
